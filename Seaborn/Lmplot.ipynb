```python
import seaborn as sns
%matplotlib inline
#load seaborn builtin data set
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')#scatter plot with a linear fit on top, adding hue means theres two scatter plot and linear fit
```

```python
#using markers
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'])
```
