```python
import seaborn as sns
%matplotlib inline
```

```python
#using seaborn built in datasets available to load
tips = sns.load_dataset('tips')
tips.head()
```

```python
#dist plot - allows to show the distribution of a uni variant set of observation, univariant is way of saying one variable
sns.distplot(tips['total_bill'], kde=False, bins=40) # checking how the total bill is distributed
#kde(kernal density estimation plot) plot can be removed, leaving me with a histogram shwoing prices ranging from $10-20 then start to see a drop
```

```python
#3 joingplot- allows you to combine/compare distributions, like the one above

sns.jointplot(x='total_bill',y='tip', data=tips)

#as you can see, when you go higher in total_bill, you go higher in tip
# use kind parameter to change scatter plot to hexagon 
sns.jointplot(x='total_bill',y='tip', data=tips, kind='hex')
#or regression
sns.jointplot(x='total_bill',y='tip', data=tips, kind='reg')
```

```python
#4 pair plot - plot pair wise relationships accross entire df(it will basically do a jointplot of everyposible combination of the numerical column)
sns.pairplot(tips)
#hue allows you to pass in column name of a categorical column, e.g, sex column is categorical(male, female)
sns.pairplot(tips, hue='sex')
```

```python
#5 redrug 
sns.rugplot(tips['total_bill'])
```
