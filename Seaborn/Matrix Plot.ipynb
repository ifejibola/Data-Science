```python
import seaborn as sns
%matplotlib inline
#loading seaborn built-in dataset
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
#heatmaps primary way of showing matrix plot,in order to use a heatmap : data should alreayd be in matrix form
#index and column name match up so cell value indicate data relavant to both index and column

#corrolation data
tc = tips.corr()#matrix form of data, column and index now show corrolation/relevance
tc
```

```python
sns.heatmap(tc, annot = True, cmap='coolwarm')
```

```python
flights
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
fp
```

```python
#heatmap to visualize data
sns.heatmap(fp, cmap='magma', linecolor='white', linewidth=1)
#as you can see from the data more flights occur over the years by 1960
```

```python
#Cluster Map - allows you to show relationship between data values, 1949-1950 had similar range in flights
sns.clustermap(fp)
```
