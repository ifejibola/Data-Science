```python
#1
myDF = pd.read_csv('exam.csv')
myDF.head(3)
rangeMean = myDF['Range'].mean()
print(rangeMean)
myDF[myDF['Type'] == 'Z']['Type'].count()

myDF['Type'] = myDF['Type'].replace('Z', 'XYZ', regex=True)
myDF.head()
myDF[myDF['Type'] == 'XYZ']['Type'].count()
```
